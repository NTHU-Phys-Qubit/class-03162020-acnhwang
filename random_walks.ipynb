```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# 1D random walk with step in range of [-0.5, 0.5]

number_of_results = 10000
number_of_steps = 1000
origin = [0]
y = np.array([])
y_i = 0

steps = np.random.random(number_of_steps) - 0.50
path = np.concatenate([origin, steps]).cumsum(0)
stop = path[-1:]
y = np.append(y, stop)

for i in range(number_of_results - 1):
    x_i = np.random.random(number_of_steps) - 0.50
    for j in range(number_of_steps):
        y_i = y_i + x_i[j]
    y = np.append(y, y_i)
    y_i = 0


plt.figure(figsize=(16,8))
plt.scatter(np.arange(number_of_steps + 1), path)
plt.title('1D random walk of 1 result among 10000 results', fontsize = 16)

fix, ay = plt.subplots(figsize = (8,8))
_ = ay.hist(y, bins=50)
plt.title('Distribution of 10000 results in 1D random walk from -0.5 to 0.5', fontsize = 16)
```

```python
# 1D random walk with step = -1 or 1

number_of_results = 10000
number_of_steps = 1000
step_set = [-1,1]
origin = [0]
y = np.array([])
y_i = 0
path = np.array([])

steps = np.random.choice(a = step_set, size = number_of_steps)
path = np.concatenate([origin, steps]).cumsum(0)
stop = path[-1:]
y = np.append(y, stop)

for i in range(number_of_results-1):
    steps_i = np.random.choice(a = step_set, size = number_of_steps)
    for j in range(number_of_steps):
        y_i = y_i + steps_i[j]
    y = np.append(y, y_i)
    y_i = 0
    
plt.figure(figsize=(16,8))
plt.title('1D random walk of 1 result among 10000 results', fontsize = 16)
plt.scatter(np.arange(number_of_steps + 1), path)

fix, ax = plt.subplots(figsize = (8,8))
_ = ax.hist(y, bins = 50)
plt.title('Distribution of 10000 results in 1D random walk with step = -1 or 1', fontsize = 16)
```
